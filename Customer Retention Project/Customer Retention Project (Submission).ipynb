## What the data says,

Customer satisfaction has posited as a key stimulant of purchase and repeat purchase of the produce that is called Customer Retention. Five major factors that contributed to the success of an e-commerce store have been identified as-  Service quality, System quality, Information quality, Trust and Net benefit.

The repeat purchase intention or say, loyalty of the customer gets positively affected by two factors namely, Hedonic values and Utilitarian values. Utilitarian value is proposed as a formative second-order construct formed by product offerings, product information, monetary savings and convenience. Gratification, Role, Best Deal, Social Status and Adventure are the ones represents the Hedonic values.

In the given dataset, there are various columns that falls in above five factors which in turn falls under the Hedonic and Utilitarian values that Postively affects the Customer retention. Using the given dataset we are going to analyze what is exactly customer inputs are depicting about the customer's intention of repurchase.

## Customer Retention based on Hedonic and Utilitarian values :

There are two main categories under which some columns from the given dataset falls i.e. Hedonic values and Utilitarian values. Hedonic values and Utilitarian values both combinedly affects the customers intention of repeat purchase of the any product. 



>>  HEDONIC VALUES :- shows that how the consumer experienced (i.e. pleasure, fun, and enjoyment, etc.) from buying the product. There are five columns in the dataset represents the hedonic values by the customers as follows-

(1)'45 You feel gratification shopping on your favorite e-tailer' = it shows customers Gratification after buying the product. 
(2)'46 Shopping on the website helps you fulfill certain roles' = it shows various Roles fulfilled by customers after buying the product.

(3)'47 Getting value for money spent' = it shows Best deals they get on purchasing any product.

(4)'44 Shopping on your preferred e-tailer enhances your social status' = it shows how the customers social status enhaced.

(5)'43 Shopping on the website gives you the sense of adventure' = it shows how the purchasing of any product gave the sense of adventure to the customer.




>> UTILITARIAN VALUES :- shows the products purchased by the customers are depends upon his/her needs and shows their practical uses of those products. There are four columns in the dataset represents the Utilitarian values by the customers as follows-


(1)'39 Offering a wide variety of listed product in several category' = this column shows the Offerings on different products.

(2)'42 The Convenience of patronizing the online retailer' = this columns shows the convenience of using the website for purchasing the product by the customer.

(3)'40 Provision of complete and relevant product information' = this column gives how to customer found the given information on the website relevant for their desired product.

(4)'41 Monetary savings' = this column gives the info about how the customer able to get some monetary savings from purchasing the product.

## Correlation :

#### We are creating a new dataframe to find the correlation between Hedonic & Utilitarian values and various attributes like Gender,  Age 

We are making a dataframe here which is comprised of both Hedonic values and Utilitarian values which affects positively on the intention of the customer to re-purchase the products.

```python
df_new = df[['45 You feel gratification shopping on your favorite e-tailer','46 Shopping on the website helps you fulfill certain roles','47 Getting value for money spent','44 Shopping on your preferred e-tailer enhances your social status','43 Shopping on the website gives you the sense of adventure','39 Offering a wide variety of listed product in several category','42 The Convenience of patronizing the online retailer','40 Provision of complete and relevant product information','41 Monetary savings','1Gender of respondent','2 How old are you? ']]
```

```python
df_new
```

In the above dataframe, first five columns represents the Hedonic values and four columns after that shows the Utilitarian values. We added Gender and Age columns in the dataframe to find out the correlation between different Genders and different Age groups with respect to these values.

```python
df_new.shape
```

```python
df_new.corr()
```

```python
corr = df_new.corr()
plt.subplots(figsize=(15,10))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

As the heatmap shows, the intersection of the age and social status column (hedonic value) shows the highly negative correltion between them it means when one of the variable increases the other decreases i.e. as the age of the consumer increases the social status gained by shopping from the e-tailer will decrease and vice-versa.

Also, the correlation between the Age and Gratification column is also negatively correlated.

```python
#Check correleation between the variables using Seaborn's pairplot. 
sns.pairplot(df_new, hue='1Gender of respondent', palette='CMRmap_r')
```
